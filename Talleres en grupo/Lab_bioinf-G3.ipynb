```python
def funcion_creada_por_el_grupo_3(target,problema):
  problema = list(problema) #Conversión a lista, problema = secuencia que se quiere buscar.
  target = list(target) #Conversión a lista, target = secuencia en la cual se busca la secuencia(problema).
  p=len(problema) #Conteo de los items de problema.
  l=0 #Variable para medir posición del item en la lista(problema).
  s=0 #Es un contador de la posición de la lista(target).
  for m in target: #comienza el conteo de los item en target.
    if l == p: #Condición para probar si el la secuencia que se busca ya se encontro por completo(p).
      print(f"Posición final{(s-p),s}") #Imprime el primera (s-cantidad de item(p)) y última(s) posición de la secuencia que se busca.
      break #Fin de la ejecución del condigo si de cumple la condición.
    i = problema[l] #Primer item de la primera posición de la secuencia(problema).
    if m == i: #Condición que comienza a buscar la secuencia(problema).
      if m == problema[l]: #Comprueba si item de target es igual a el item de problema.
        s+=1 #Se suma 1 a contador para seguir contando la posición.
        l+=1 #Suma uno para cambiar al siguiente item de la secuencia(problema).
        continue  #Se devuelve para seguir comprobando la secuencia(problema).
    else: #Condición cuando la comprobación es incorreta.
      l-=l #Se resta la misma cantida de posiciones que se llevaba en la comprobación de la secuencia(problema).
      s+=1 
      continue #Se devuelve para seguir buscando la secuencia(problema).
  print("Fin de busqueda") #Fin del código.
```
