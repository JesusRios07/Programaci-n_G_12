1.Cree una imagen de un canal, es decir con una forma (alto, ancho) del tamaño que desee, donde la diagonal principal hacia arriba sea blanca y desde la diagonal hacia abajo sea negra.

```python
import numpy as np
from skimage import data
import matplotlib.pyplot as plt
```

```python
ancho=100
imagen=[]
pixel_blanco=[255,2555,255]
pixel_negro=[0,0,0]
for i in range (0,ancho):
  fila=[]
  for j in range(0,ancho):
    if i>=j :
      fila.append(pixel_negro)
    else :
      fila.append(pixel_blanco)
  imagen.append(fila)
imagen_negra=np.array(imagen)
plt.imshow(imagen_negra)
print(imagen_negra)
```

2.Use una imagen de la librería skimage, en la base de datos “data”.

```python
from skimage import filters
from skimage.color import rgb2gray
import matplotlib.pyplot as plt
```

```python
imagen = data.cat()
plt.imshow(imagen)
plt.show
imagen_g=rgb2gray(imagen)#rgb2gray nos permite convertir una imagen con de terminados canales de rgb a escalas de  grises
filtros=[filters.sobel, filters.roberts, filters.prewitt]#Definimos la variable filtros
for filtro in filtros:
  img_fil=filtro(imagen_g)#Aplicamos cada uno de los filtros a la imagen seleccionada
#Mostramos los resultados 
plt.imshow(img_fil)
plt.show()
```
